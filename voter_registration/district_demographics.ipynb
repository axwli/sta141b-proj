```python
import pandas as pd
```

```python
reg_2016 = pd.read_csv('2016_registration.csv')
reg_2020 = pd.read_csv('2020_registration.csv')
reg_both = pd.read_csv('registration_combined.csv')
district_2020 = pd.read_csv('nc_district_2020.csv')
district_2020.head()
```

# 2020 

```python
# making the county names all uppercase to match 
district_2020['County'] = district_2020['County'].str.upper()
```

```python
reg_2020
# adding non hispanic population
reg_2020['non_hispanic'] = reg_2020['total'] - reg_2020['hispanic']

column_to_move = reg_2020.pop('non_hispanic')  
reg_2020.insert(9, 'non_hispanic', column_to_move)
reg_2020
```

```python
# merging on county
demo_2020 = pd.merge(district_2020, reg_2020, how='left', left_on = 'County', right_on ='index').drop(columns=['index', 'date'])
#demo_2020[demographics].multiply(demo_2020['Percent of District Pop in County']/100, axis=0)
demo_2020.head()
```

```python
# column names of the demographic columns 
demo_groups = demo_2020.columns[7:].tolist()
demo_2020[demo_groups] = demo_2020[demo_groups].multiply(demo_2020['Percent of County Pop in District']/100, axis=0).round(0)
demo_2020.head(5)
```

```python
# column which contains list of counties for each district
district_counties = demo_2020.groupby('District')['County'].apply(lambda x: list(x.unique())).reset_index()
district_counties
```

```python
# getting sums across demographic fielst
demo_drop = demo_2020.drop(['County', 'Total County Pop', 'Total District Pop','Percent of County Pop in District', 'Percent of District Pop in County'], axis = 1)
districts_totaled = demo_drop.groupby('District').sum()
districts_totaled
```

```python
district_demo_2020 = pd.merge(districts_totaled, district_counties, how='left', on= 'District')
district_demo_2020.rename(columns={'County Pop in District': 'district_pop', 'total':'n_reg_voters'}, inplace=True)

district_demo_2020
```

```python
demo_groups
```

```python

# adding column for percentage of each demographic over n_reg voters 
for col in demo_groups[1:]:
    district_demo_2020[f'{col}_percent_of_reg_voters'] = district_demo_2020[col] / district_demo_2020['n_reg_voters']

for col in demo_groups[1:]:
    district_demo_2020[f'{col}_percent_of_pop'] = district_demo_2020[col] / district_demo_2020['district_pop']

district_demo_2020
```

```python
district_demo_2020.to_csv('district_demo_2020.csv', index=False) 
```

# 2016

```python
district_2016 = pd.read_csv('nc_pop_2016.csv')
district_2016['County'] = district_2016['County'].str.upper()
# turning country district pop into a int since it will be used later in a calculation 
district_2016['County District Pop'] = district_2016['County District Pop'].str.replace(",", "").astype(float)

district_2016.head()
```

```python
reg_2016['non_hispanic'] = reg_2016['total'] - reg_2016['hispanic']

column_to_move = reg_2016.pop('non_hispanic')  
reg_2016.insert(9, 'non_hispanic', column_to_move)
reg_2016
```

```python
# merging on county
demo_2016 = pd.merge(district_2016, reg_2016, how='left', left_on = 'County', right_on ='index').drop(columns=['index', 'date'])
demo_2016.head(10)
```

```python
# column names of the demographic columns 
demo_groups = demo_2016.columns[7:].tolist()
demo_2016[demo_groups] = demo_2016[demo_groups].multiply(demo_2016['Percent County in District']/100, axis=0).round(0)
demo_2016.head(5)
```

```python
# column which contains list of counties for each district
district_counties = demo_2016.groupby('District')['County'].apply(lambda x: list(x.unique())).reset_index()
district_counties
```

```python
# getting sums across demographic fielst
demo_drop = demo_2016.drop(['County', 'County Pop', 'District Pop','Percent County in District', 'Percent District in County'], axis = 1)
districts_totaled = demo_drop.groupby('District').sum()
districts_totaled
```

```python
district_demo_2016 = pd.merge(districts_totaled, district_counties, how='left', on= 'District')
district_demo_2016.rename(columns={'County District Pop': 'district_pop', 'total':'n_reg_voters'}, inplace=True)

district_demo_2016
```

```python
district_demo_2016.to_csv('district_demo_2016.csv', index=False) 
```
